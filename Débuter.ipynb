```python
import math
class Point:
  "Cette classe représente un point qui a des coordonnées. Et on peut calculer la distance avec d'autres ponit"
  def __init__(self, abscisse, ordonne):
    self.x = abscisse
    self.y = ordonne

  def distance(self, autre_point):
      squarred_d = ((self.x - autre_point.x)**2 + (self.y - autre_point.y)**2)
      d = math.sqrt(squarred_d)
      return(d)

  def __repr__(self):
    return f"Ce point a pour abcisse {self.x} et ordonné {self.y}"

  def origine(self):
    self.x= 0
    self.y= 0

  def projection_x(self):
    self.y = 0

  def projection_y(self):
    self.x = 0

```

```python
a = Point(2, 3)
b = Point(1, 5)
```
