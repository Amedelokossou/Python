# Exercice Numpy

```python
import numpy as np
# Math, Physique, SVT

x = np.array([[ 0, 13,  5],
       [ 5,  8,  2],
       [ 3,  7, 12],
       [ 2,  8,  0],
       [ 7,  4,  1]])
x
```

```python
#x représente les notes de 5 étudiants en Maths, physique et biologie
```

```python
#Combien de notes sont supérieur à 5 ?
notes_sup_5 = np.sum(x > 5)
notes_sup_5
```

```python
#Quel est le nombre de notes supérieur à 8 pour chaque étudiants ?
notes_sup_8_pour_chaque_etudiant = np.sum(x > 8, axis = 1 )
notes_sup_8_pour_chaque_etudiant
```

```python
#Quel est le nombre d'élèves ayant au moins 2 notes supérieures à 8 ?

nbre_etud_avec_au_moins_2_note_sup_8 = np.sum(x > 8, axis = 1 )
nbre_etud_avec_au_moins_2_note_sup_8 = np.sum(nbre_etud_avec_au_moins_2_note_sup_8 >= 2)
nbre_etud_avec_au_moins_2_note_sup_8
```

```python
#Est-ce que tous les élèves ont une note supérieur à 3 ?
nbre_etud_avec_note_sup_3 = np.sum(x > 3, axis = 1 )
nbre_etud_avec_note_sup_3
```

```python
#Est-ce qu'il y au moins un élève ayant au moins une note supérieur à 12 ?
nbre_etud_avec_au_moins_note_sup_12 = np.sum(x > 12 , axis = 1)
nbre_etud_avec_au_moins_note_sup_12
```

```python
#Quelle est la moyenne en Math et Physique de la classe si on considère uniquement les 3 élèves dont les notes apparaissent en premier ?
data = x[0:3, 0:2]
data
```
