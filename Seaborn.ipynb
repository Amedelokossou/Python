#Box plot (analyse univariée numérique)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lien = 'https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv'
```

```python
titanic = pd.read_csv(lien)
titanic
```

```python
titanic.head()
```

```python
titanic.Sex.value_counts()
```

```python
titanic.Sex.value_counts().plot(kind= 'bar')
```

```python
dir(pd.DataFrame.plot)
```

```python
titanic.Sex.value_counts().plot.pie()
```

```python
titanic.Sex.value_counts().plot(kind= 'barh')
```

```python
#Travailler avec la variable Age
sns.boxplot(data=titanic, x='Age')
```

```python
sns.boxplot(data=titanic, y='Age')
```

```python
sns.boxplot(titanic['Age'])
#sns.boxplot(data=titanic, x='Age')
```

```python
sns.kdeplot(titanic['Age'])
```

```python
sns.distplot(titanic['Age'])
```

Violin plot(x=category, y=numérique)

```python
titanic.head()
```

```python
sns.violinplot(x="Survived", y="Age", data=titanic)
```

```python
sns.countplot( x="Sex", data=titanic)
plt.title('Sexe des voyageurs')
plt.show()
```

Heatmap (plusieurs variables numériques)

```python
titanic.corr(numeric_only=True)
```

```python
sns.heatmap(titanic.corr(numeric_only=True), annot =True , cmap = 'YlGnBu')
```

```python
sns.heatmap(titanic.corr(numeric_only=True) [['Survived' , 'Age']], annot =True , cmap = 'YlGnBu')
```

Facetgrid (2 variables categorie et numérique)

```python
cond = (titanic.Sex=="male") & (titanic.Survived==0)
titanic[cond]['Age'].hist()
```

```python
grid = sns.FacetGrid(titanic, col='Sex', row='Survived')
#grid.map(plt.hist, 'Age')
```

```python
grid = sns.FacetGrid(titanic, col='Sex', row='Survived')
grid.map(plt.hist, 'Age')
```

Regplot (scatter plot du riche)

```python
sns.regplot(data=titanic, x='Age', y="Fare")
```

```python
sns.scatterplot(data=titanic, x='Age', y="Fare")
```

lmplot(facetgrid + regplot) 2 variables catégories 2 variables numériques

```python
titanic.Fare.hist()
```

```python
sns.lmplot(data=titanic, x='Age', y='Fare', row='Survived', col='Sex')
```

Strip plot : Scatter mais pour visualiser une variable numérique et une variable de categorie

```python
sns.stripplot(data=titanic, x='Survived', y='Age', jitter=True)
```

```python
sns.stripplot(data=titanic, x='Survived', y='Age')
```

Swarm plot : strip plotmais avec la distribution

```python
sns.swarmplot(data=titanic, x='Survived', y='Age')
```

```python
sns.swarmplot(data=titanic, x='Survived', y='Age', hue='Survived', legend=False)
```
