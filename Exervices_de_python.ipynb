#Exercice 1
Ecris une fonction nommée devinette qui ne prend aucun argument. Quand on l'exécute, elle choisit un nombre aléatoire en 0 et 100 (inclus)et demande à l'utilisateur de deviner ce nombre.A chaque fois que l'utilisateur entre un nombre, le programme affiche l'une de ces phrases :

Trop grand
Trop petit
Exact
Si l'utilisateur a choisit correctement, le programme s'arrete. Sinon, le programme demande encore à l'utilisateur de rentrer un nouveau nombre jusqu'à ce qu'il trouve la bonne solution et s'arrête.

Aide : Cherche le module random de Python pour générer un nombre aléatoirement.

```python
import random

def devinette():
  nombre_aleatoire = random.randint(0, 100)
  nombre_utilisateur = int(input("Deviner le nombre aléatoire entre 0 et 100 :"))
  while nombre_utilisateur != nombre_aleatoire:
    if nombre_utilisateur > nombre_aleatoire:
      print("Trop grand")
      nombre_utilisateur = int(input("Deviner un nombre plus petit  :"))
    else:
      print("Trop petit")
      nombre_utilisateur = int(input("Deviner un nombre plus grand   :"))
    if nombre_utilisateur == nombre_aleatoire:
      print("Exact")
      break
```

```python
import random

def devinette_3essaies():
    # On tire le nombre au hasard UNE SEULE FOIS, avant la boucle.
    # Il restera donc le même pendant les 3 essais.
    nombre_aleatoire = random.randint(0, 100)

    nombre_de_chances = 3  # nombre d'essais autorisés

    # La boucle se répète 3 fois : essais 1, 2 puis 3.
    for tentative in range(1, nombre_de_chances + 1):

        # On demande une proposition au joueur et on la convertit en entier.
        proposition = int(input(f"Essai {tentative} - Devine le nombre entre 0 et 100 : "))

        if proposition == nombre_aleatoire:
            print("Exact ! Bravo 🎉")
            break  # le joueur a trouvé : on sort de la boucle, le jeu est fini

        elif proposition > nombre_aleatoire:
            print("Trop grand")   # indice, mais on ne donne PAS la solution

        else:  # proposition < nombre_aleatoire
            print("Trop petit")   # indice, mais on ne donne PAS la solution

    else:
        # Ce bloc 'else' appartient au 'for'.
        # Il ne s'exécute QUE si la boucle est allée jusqu'au bout sans 'break',
        # c'est-à-dire si le joueur n'a trouvé à aucun des 3 essais.
        # C'est SEULEMENT ici qu'on révèle la bonne réponse.
        print(f"Vous avez épuisé vos {nombre_de_chances} chances. "
              f"Le nombre était {nombre_aleatoire}.")

# On lance le jeu
devinette_3essaies()
```

En Python, la fonction sum te permet de faire la somme des nombres d'une liste. Par exemple:

nombres = [5,2,6,8,9,4]

Dans cet exercice, nous allons réecrire cette fonction mais nous la nommerons somme. La fonction somme prend en entrée une liste de nombres et renvoie la somme de ces nombres. Ici, vous n'avez pas le droit d'utiliser la fonction sum de Python. Le but de cet exercice est d'apprendre à implémenter une fonction qui existe déjà et gagner en confiance sur nos capacités à faire des choses que d'autres programmeurs plus expérimentés ont déjà fait.

```python
nombres = [1, 2.0, 3, 4]
nombres_decimal = [float(x) for x in nombres]
def somme():
  total = 0.0
  i = 0
  while i < len(nombres_decimal):
    total = total + nombres_decimal[i]
    i = i + 1
  return total
print(somme())
```

Exo 8 : Début et fin des séquences
En Python, les listes, les tuples et les strings sont des séquences. En Python, on peut écrire une fonction qui prend différents types de variables. Dans cet exercice, nous allons écrire une fonction qui prend en Entrée, une liste, un tuple ou un string et retourne une variable du même type mais avec le premier et le dernier élément de la séquence.

debut_fin('deopry')
> 'dy'

debut_fin([5,2,6,9,8])
> [5, 8]

debut_fin((2,6,8,9))
> 2, 9

```python
sequence = input("Entrez une chaine de caractères, une liste ou un tuple : ")
def debut_fin(sequence):
    # Si la séquence est vide, on la renvoie telle quelle
    if len(sequence) == 0:
        return sequence

    # On récupère le premier élément (indice 0) et le dernier (indice -1)
    premier = sequence[0]
    dernier = sequence[-1]

    # Si c'est un texte (string)
    if type(sequence) == str:
        return premier + dernier

    # Si c'est une liste
    elif type(sequence) == list:
        return [premier, dernier]

    # Si c'est un tuple
    elif type(sequence) == tuple:
        return (premier, dernier)
print(debut_fin(sequence))
```

Exo 11 : Distance entre plusieurs points
Ecrivez une fonction qui prend une liste des nombres et qui retourne une liste qui contient la distance entre les différents nombres successifs.

Exemple

input : [65, 54, 59]
output : [11, 5]

```python
import math
#liste_de_nombre = input("Entrez une liste de nombre séparé par des espaces : ")
#liste_de_nombre = [int(x) for x in liste_de_nombre.split()]
liste_de_nombre = [65, 54, 59]

class Point:
  "Cette classe représente un point qui a des coordonnées. Et on peut calculer la distance avec d'autres point"
  def __init__ (self, abscisse):
    self.x = abscisse

  def distance(self, j):
    squarred_d = ((self.x - j.x)**2)
    d = math.sqrt(squarred_d)
    return(d)

for index_i, i in enumerate(liste_de_nombre):
  for j in liste_de_nombre[index_i + 1:]:
    if i != j:
      point_i = Point(i)
      point_j = Point(j)

      distance_entre_point = point_i.distance(point_j)
      print(f'la distance entre les points {point_i.x} et {point_j.x} est de {distance_entre_point} ')
```

```python
menu = {'riz' : 4500.50 , 'atassi' : 2500, 'pate' : 3000.90}
def restaurant():
  total = 0.0

  print(f'Bienvenu dans notre Restaurant !' )
  while True:
    commande = str(input('Votre Commande :'))

    if commande == "":
      print(f'Votre total est de {total}')
      break
    elif commande in menu.keys():
      prix_commande = float(menu[commande])
      total = total + prix_commande
      print(f'Le prix de  {commande} est de : {prix_commande} FCFA. Le total actuel de vos commandes est de {total}')

    else:
      print(f'La commande  {commande} exite pas.')

restaurant()
```
